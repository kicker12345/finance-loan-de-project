```python
spark.sql("""
CREATE OR REPLACE VIEW finance_project.customers_loan_v AS 
SELECT
    l.loan_id,
    c.member_id,
    c.emp_title,
    c.emp_length,
    c.home_ownership,
    c.annual_income,
    c.address_state,
    c.address_zip_code,
    c.address_country,
    c.grade,                    -- Fixed: from customers_external
    c.sub_grade,                -- Fixed: from customers_external  
    c.verification_status,      -- Check if this exists in loans_external
    c.tot_hi_cred_limit,
    c.application_type,
    c.join_annual_income,
    c.verification_status_joint,
    l.loan_amount,
    l.funded_amount,
    l.loan_term_years,
    l.interest_rate,
    l.monthly_installment,
    l.issue_date,
    l.loan_status,
    l.loan_purpose,
    r.total_principal_received,
    r.total_interest_received,
    r.total_late_fee_received,
    r.last_payment_date,
    r.next_payment_date,
    d.delinq_2yrs,
    d.delinq_amnt,
    d.mths_since_last_delinq,
    e.pub_rec,
    e.pub_rec_bankruptcies,
    e.inq_last_6mths
FROM finance_project.customers_external c
LEFT JOIN finance_project.loans_external l ON c.member_id = l.member_id
LEFT JOIN finance_project.loans_repayments_external r ON l.loan_id = r.loan_id
LEFT JOIN finance_project.loans_defaulters_delinq_external d ON c.member_id = d.member_id
LEFT JOIN finance_project.loans_defaulters_detail_records_enq_external e ON c.member_id = e.member_id
""")
```
